```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sam_data = pd.read_csv("survey_results_public.csv")
sam_data.head()
```

```python
sam_data.describe()
```

```python
sam_data.tail(6)
```

```python
label = "ConvertedCompYearly"

#Drop rows with no data
sam_data = sam_data.dropna(subset = [label])

#drop rows with extreme outliers
sam_data = sam_data.drop(sam_data[sam_data[label] > 40000].index)

#checking if the number make sense
sam_data[[label]].describe()
```

```python
numeric_feature = ["YearsCode", "YearsCodePro"]
for col_name in numeric_feature:
    sam_data[col_name] = pd.to_numeric(sam_data[col_name], errors = "coerce")
    sam_data = sam_data.dropna(subset= [col_name])
sam_data[numeric_feature].describe()
```

```python
label_data = sam_data[label]
fig = plt.figure(figsize=(4, 2))
ax = fig.gca()
ax.hist(label_data, bins=100)
ax.set_ylabel("Frequency")
ax.axvline(label_data.mean(), color="red", linestyle = "dashed", linewidth = 2)
ax.axvline(label_data.median(), color="green", linestyle = "dashed", linewidth = 2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=len(numeric_feature), figsize = (10,4))

for ind, col_name in enumerate(numeric_feature):
    feature = sam_data[col_name]
    axis = axes[ind]
    feature.hist(bins= 100, ax = axis)
    axis.axvline(feature.mean(), color="black", linestyle = "dashed", linewidth = 2)
    axis.axvline(feature.median(), color="red", linestyle = "dashed", linewidth = 2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=len(numeric_feature), figsize = (10,4), sharey= True)

for ind, feature in enumerate(numeric_feature):
    label_data = sam_data[label]
    feature_data = sam_data[feature]
    correlation = feature_data.corr(label_data)
    axis = axes[ind]
    axis.scatter(x= feature_data, y = label_data)
    axis.set_xlabel(feature)
    axis.set_ylabel(label)
    axis.set_title(f'{label}vs{feature}\n Correlation:{correlation}')
```

```python
#Separating features from labels

x = sam_data[numeric_feature].values
y = sam_data[label].values

print('feature: ', x[:5], '\nlabels:', y[:5], sep='\n')
```

```python
#splitting the data

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=0)

print (f'Training Set: {x_train.shape[0]} rowws\n Test Sets : {x_test.shape[0]} rows')
```

```python
#plotting a regression model

from sklearn.linear_model import LinearRegression

Model = LinearRegression().fit(x_train, y_train)

print(Model.coef_)
print(Model.intercept_)

#interpretation
# y= mx + c
# y = 464.18x  -113.1x + 16744.4
```

```python
Predictions = Model.predict(x_test)
np.set_printoptions(suppress=True)

print('Predicted labels :', np.round (Predictions) [:8])

print('Actual labels :', y_test[:8])
```

```python
Predictions = Model.predict(x_test)
np.set_printoptions(suppress=True)

print('Predicted labels :', np.round (Predictions) [:8])

print('Actual labels :', y_test[:8])
```

```python
plt.scatter(y_test, Predictions)
plt.xlabel(f'Actual{label}')
plt.ylabel(f'Predicted{label}')

#regression line
a = np.polyfit(y_test, Predictions, 1)
b = np.poly1d(a)
plt.plot(y_test, b(y_test), color = 'red')
```
